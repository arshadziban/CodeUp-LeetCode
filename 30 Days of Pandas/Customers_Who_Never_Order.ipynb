```python
import pandas as pd
```

```python
data = [[1, 'Joe'], [2, 'Henry'], [3, 'Sam'], [4, 'Max']]
customers = pd.DataFrame(data, columns=['id', 'name']).astype({'id':'Int64', 'name':'object'})


data = [[1, 3], [2, 1]]
orders = pd.DataFrame(data, columns=['id', 'customerId']).astype({'id':'Int64', 'customerId':'Int64'})
```

```python
customers
```

```python
orders
```

```python
no_order_customers = customers[~customers['id'].isin(orders['customerId'])]
```

In your code:

`customers['id'].isin(orders['customerId'])` gives **True** for customers who **made orders**.

`~customers['id'].isin(orders['customerId'])` gives **True** for customers who **did NOT make orders**.

That’s why we use `~` — we want the ones who did **NOT** order.


```python
no_order_customers
```

```python
order_customers = customers[customers['id'].isin(orders['customerId'])]
```
